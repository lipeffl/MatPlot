##Scatterplot

```python
#scatterplot
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 5, 6, 8, 4, 8] 
#usando o marker:
'''
'o' = bolinha 
'i' = |
'v' = triangulo
'x' = x #marker, para definir 'ponteiro' do gráfico
'''
plt.scatter(x, y, label =  'Pontos', color = 'r', marker = 'x', s = 80)
plt.legend()
plt.show()
```

##Stack Plots

```python
#stack plots
dias = [1, 2, 3, 4, 5]
dormir = [7, 8, 6, 7, 7]
comer = [2, 3, 4, 5, 3] 
trabalhar = [7, 8, 7, 2, 2] 
passear = [8, 5, 7, 8, 13] #passear não aparece no gráfico

plt.stackplot(dias, dormir, comer, trabalhar, passear, labels = ['dias', 'dormir', 'comer', 'trabalhar', 'passear'])
plt.legend(loc = 'upper left')
plt.show()
```

```python
#controle dos eixos
x = np.linspace(0, 1., 100)
fig, axes = plt.subplots(1, 2, figsize = (12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Eixos com arange padrão")

x = np.linspace(-1, 1., 100) 
axes[1].plot(x, x**2, x, x**3)
axes[1].set_ylim([0, 1])  #especifico tamanho de y
axes[1].set_xlim([-1, 1]) #especifico tamanho de x  #tetar diferentes tamanhos
axes[1].set_title("Eixos customizados")
```

```python
xx = np.linspace(-0.75, 1., 100)
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize = (12, 3))
axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")
#arescentado - segundo grafico - tipo step
axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")
```

```python
xx = np.linspace(-0.75, 1., 100)
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize = (12, 3))
axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")
axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")
#acrescentado terceiro grafico - tipo bar
axes[2].bar(n, n**2, align = "center", width = 0.5, alpha = 0.5)
axes[2].set_title("bar")
```

```python
xx = np.linspace(-0.75, 1., 100)
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize = (12, 3))
axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")
axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")
axes[2].bar(n, n**2, align = "center", width = 0.5, alpha = 0.5)
axes[2].set_title("bar")
#acrescentado - quarto grafico
axes[3].fill_between(x, x**2, x**3, color = "green", alpha = 0.5)
axes[3].set_title("fill_between")
```

##Histogramas

```python
#histogramas
n = np.random.randn(100)
fig, axes = plt.subplots(1, 2, figsize = (12, 4)) #colocando outros valores para aumentar tamanho do gráfico

axes[0].hist(n)
axes[0].set_title("Histograma padrão") #titulo
axes[0].set_xlim((min(n)), (max(n)))

```

```python
#histogramas
n = np.random.randn(100)
fig, axes = plt.subplots(1, 2, figsize = (12, 4)) #colocando outros valores para aumentar tamanho do gráfico

axes[0].hist(n)
axes[0].set_title("Histograma padrão") #titulo
axes[0].set_xlim((min(n)), (max(n)))
axes[1].hist(n, cumulative = True, bins = 50) #axes 1 segundo grafico
axes[1].set_title("Histograma cumulativo")
axes[1].set_xlim((min(n), max(n)))
```

```python
#mapa de calor(color map)
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def ColorMap(phi_m, phi_p):
    return ( + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2 * phi_p))


phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
x, y = np.meshgrid(phi_p, phi_m)
z = ColorMap(x, y).T

fig, ax = plt.subplots()

p = ax.pcolor(x/(2*np.pi), y/(2*np.pi), Z, vmin = abs (Z).min(), vmax = abs(Z).max()) #cmap=cm. 
cb = fig.colorbar(p, ax = ax)
```

##Gráficos 3D

```python
#gráficos 3d
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
fig = plt.figure(figsize = (14,6))

# Um objeto Axes3D é criado como qualquer outro eixo usando a palavra-chave: projection='3d'
ax = fig.add_subplot(1, 2, 1, projection = '3d')
p = ax.plot_surface(x, y, z, rstride = 4, cstride = 4, linewidth = 0)

ax = fig.add_subplot(1, 2, 2, projection = '3d')
p = ax.plot_surface(x, y, z, rstride = 1, cstride = 1, cmap = plt.cm.coolwarm, linewidth = 0, antialiased = False)
cb = fig.colorbar(p, shrink = 0.5)
```

```python
#wire frame
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1, 1, 1, projection = '3d')
p = ax.plot_wireframe(x, y, z, rstride=4, cstride=4)
```

```python
fig = plt.figure(figsize = (8, 6))

ax = fig.add_subplot(1, 1, 1, projection = '3d')

ax.plot_surface(x, y, z, rstride=4, cstride=4, alpha=0.25)
cset = ax.contour(x, y, z, zdir = 'z', offset = -np.pi, cmap = plt.cm.coolwarm)
cset = ax.contour(x, y, z, zdir = 'x', offset = -np.pi, cmap = plt.cm.coolwarm)
cset = ax.contour(x, y, z, zdir = 'y', offset = 3*np.pi, cmap = plt.cm.coolwarm)

ax.set_xlim3d(-np.pi, 2*np.pi);
ax.set_ylim3d(0, 3*np.pi);
ax.set_zlim3d(-np.pi, 2*np.pi);
```
